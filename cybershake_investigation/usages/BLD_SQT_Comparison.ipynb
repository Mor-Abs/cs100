# Import dependencies

```python
import os
import yaml
from pathlib import Path
import pandas as pd
```

# Path and directory works

```python
file_path = Path.cwd()
root_path = file_path.parent.parent
data_path = root_path / "Shared_BLD_SQT_Data"
```

# Define versions and faults

```python
vers = ["v21p6", "v23p5"]
faults = ["Esk", "Hanmer", "Poulter", "Waitohi", "WhiteCk"]
```

# Global functions

# Load data

```python
REL_DATA = {f: None for f in faults}
for f in faults:
    vers = [dir.stem.split("_")[2] for dir in data_path.iterdir() if dir.is_dir() and dir.stem.split("_")[0] == f]
    REL_DATA[f] = {v: None for v in vers}
    for v in vers:
        directory = data_path / f"{f}_Source_{v}"
        Rels = [item.stem for item in directory.rglob("*_REL*.csv") if not item.name.endswith(".pertb.csv")]
        Rels.sort()
        data_df = pd.DataFrame()
        for rl in Rels:
            file = directory / f"{rl}.csv"
            temp_df = pd.read_csv(file)
            data_df[rl] = temp_df.iloc[0]
        REL_DATA[f][v] = data_df.T
            
            
        print(Rels)
        # for item in directory.iterdir():
        #     print(f"{item}")
            
    

    
```
